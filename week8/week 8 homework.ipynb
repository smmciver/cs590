```r
#Import peanut_lines.csv and store data as lines
#Show the first 25 rows of lines

lines<-read.csv('peanut_lines.csv')
head(lines, 25)
```

```r
#Create a new column called 'Bailey_Parent' where the value is 'Yes' if 'Bailey' was used in the development of that line, 
#otherwise use 'No'.
#Show the first 10 rows of lines.

Bailey <- function(x){
    if (grepl('Bailey',x)){
        return('Yes')
    }else{
        return('No')
    }
}

```

```r
lines$Bailey_Parent<-sapply(lines$Parentage,Bailey)
head(lines, 10)
```

```r
(lines)
```

```r
#What percentage of lines were developed using 'Bailey' are verified as high-oleic (Fatty Acid Genotype [FAG] = 'ol ol')? 
#Round your answer to the tenth. Do not include the 'Check' lines.
#Note: Bailey (FAG) = '+ +', FAG = 'Check', means it hasn't been identified as high-oleic yet.

bailey_percent<-round((length(which((grepl('ol ol', lines$FAG) & grepl('Bailey',lines$Parentage))==TRUE))/nrow(lines))*100, 1)
```

```r
sprintf('%s%%', bailey_percent)
```

```r
#Replace the name 'Bailey' with the tested name 'N03081T' permanently for all instances in the 'Parentage' column.
#Show the first 10 lines


lines$Parentage<-gsub('Bailey', 'N03081T', lines$Parentage)
head(lines, 10)
```

```r
#Load the 'peanut_yield.txt' dataset as 'yields'
#Merge (right join, all.y=TRUE) 'lines' and 'yields' together on 'NC_Accession' and name it 'peanut_db'


yields <- read.csv('peanut_yield.txt', header=TRUE, sep='\t')
```

```r
head(yields, 20)
```

```r
peanut_db <- merge(lines, yields, by.x='NC_Accession', by.y='NC_Accession', all.y=TRUE)
```

```r
head(peanut_db)
```

```r
peanut_db <- peanut_db[peanut_db$NC_Accession == 'Bailey' | peanut_db$Bailey_Parent == 'Yes',]
```
